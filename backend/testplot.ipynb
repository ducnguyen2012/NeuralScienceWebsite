```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import torch 
```

```python
def generateRandomSample():
    '''
    This function will generate random signal with shape <1,1,22,1001>
    '''
    n_samples = 1  # We want to generate a single sample
    n_channels = 22  # 22 EEG channels
    n_timepoints = 1001  # Time points

    # Randomly generate EEG data (values typically in the range of -100 to 100 ÂµV)
    randomSample = np.random.randn(n_samples, 1, n_channels, n_timepoints).astype(np.float32)  # Single channel, 22 channels, 1001 time points

    # Convert the random sample to a PyTorch tensor
    randomSampleTensor = torch.from_numpy(randomSample)
    return randomSampleTensor
```

```python
def plotData(sample):
    eeg_data = sample.squeeze().numpy()  # Remove dimensions of size 1 to get (22, 1001)
    plt.figure(figsize=(15, 20))  # Adjust figure size
    for channel in range(eeg_data.shape[0]):  # Loop over 22 channels
        plt.subplot(11, 2, channel + 1)  # Create a 11x2 grid of subplots
        plt.plot(eeg_data[channel])  # Plot the data for this channel
        plt.title(f'Channel {channel + 1}')  # Add title
        plt.xlabel('Time Points')
        plt.ylabel('Amplitude (ÂµV)')
        plt.tight_layout()  # Adjust subplot spacing
    plt.show()
plotData(generateRandomSample())
```
